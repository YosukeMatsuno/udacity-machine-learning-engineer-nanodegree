**Result**: Next day close price. (true price for prediction)

### [ LTSM RNN (main) ]

### Split training-test and validation dataset

```python
ohlc_train_test = nikkei_ohlc.copy()
ohlc_train_test = ohlc_train_test["2001" : "2015"]
ohlc_validation = nikkei_ohlc.copy()
ohlc_validation = ohlc_validation["2015" : "2018"]
```

### 1. Normalization and clean dataset 

```python
mean = ohlc_train_test['Close'].mean()

ohlc_train_test['Close'] = ohlc_train_test['Close'] / mean
ohlc_train_test['Result'] = ohlc_train_test['Result'] / mean
ohlc_train_test['Variation'] = ohlc_train_test['Variation'] / mean
ohlc_train_test = ohlc_train_test.drop(["Open", "High", "Low", "Volume", "Variation_rate"], axis= 1)

ohlc_validation['Close'] = ohlc_validation['Close'] / mean
ohlc_validation['Result'] = ohlc_validation['Result'] / mean
ohlc_validation['Variation'] = ohlc_validation['Variation'] / mean
ohlc_validation = ohlc_validation.drop(["Open", "High", "Low", "Volume", "Variation_rate"], axis= 1)
```

### 2. make dataset: (n) days historical price list and next day price

```python
import numpy as np

def generate_lstm_dataset(ohlc, period_day, overlap= False):

    stock_price = []
    result_price = []

    # avoid overlap
    if overlap == False:
        for index in range(0, len(ohlc)-period_day, period_day):
            stock_price.append(ohlc["Close"][index : index+period_day].values)
            result_price.append(ohlc["Close"][index+period_day])
    else:
        for index in range(len(ohlc)-period_day):
            stock_price.append(ohlc["Close"][index : index+period_day].values)
            result_price.append(ohlc["Close"][index+period_day])

    stock_price = np.array(stock_price)
    
    return {"stock_price": stock_price, "result_price": result_price}
```

```python
period_day = 30

LSTM_dataset = generate_lstm_dataset(ohlc_train_test, period_day)
stock_price, result_price = LSTM_dataset["stock_price"], LSTM_dataset["result_price"]
```

```python
print ("stock_price: {} result_price: {}".format(len(stock_price), len(result_price)))
```

### 3. Split training and test data set

```python
train_size = 0.8

X_train = stock_price[0 : int(train_size*len(stock_price))]
y_train = result_price[0 : int(train_size*len(result_price))]

X_test = stock_price[int(train_size*len(stock_price)) :]
y_test = result_price[int(train_size*len(result_price)) :]

# adjust shape
a, b = X_train.shape
X_train = X_train.reshape(a, b , 1)

a, b = X_test.shape
X_test = X_test.reshape(a, b , 1)
```

### 4. Build LSTM model

```python
from keras.models import Sequential  
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM

in_out_neurons = 1
hidden_neurons = 200
length_of_sequences = period_day

LSTM_model = Sequential()

LSTM_model.add(LSTM(hidden_neurons, batch_input_shape= (None, length_of_sequences, in_out_neurons), return_sequences= False))
LSTM_model.add(Dropout(0.2))
LSTM_model.add(Dense(in_out_neurons))
LSTM_model.add(Activation("linear"))
```

```python
LSTM_model.compile(loss= "mean_squared_error", optimizer= "adam")
history = LSTM_model.fit(X_train, y_train, batch_size= 600, epochs= 100, validation_split= 0.2)
```

### 5. Learning Curve

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='valid')
plt.title("Learning Curve")
plt.legend()
```

### 6. Predict

```python
y_pred = LSTM_model.predict(X_test) 
```

### 7. Evaluation

```python
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

### 3. Split training and test data set

```python
train_size = 0.8

X_train = stock_price[0 : int(train_size*len(stock_price))]
y_train = result_price[0 : int(train_size*len(result_price))]

X_test = stock_price[int(train_size*len(stock_price)) :]
y_test = result_price[int(train_size*len(result_price)) :]

# adjust shape
a, b = X_train.shape
X_train = X_train.reshape(a, b , 1)

a, b = X_test.shape
X_test = X_test.reshape(a, b , 1)
```

### 4. Build LSTM model

```python
import os
import time
import warnings
import numpy as np
from numpy import newaxis
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.models import Sequential
warnings.filterwarnings("ignore")

LSTM_model_v2 = Sequential()
LSTM_model_v2.add(LSTM(input_shape= (period_day, 1),
               output_dim= 100,
               return_sequences= True))

LSTM_model_v2.add(Dropout(0.2))
LSTM_model_v2.add(LSTM(100, return_sequences= False))
LSTM_model_v2.add(Dropout(0.2))
LSTM_model_v2.add(Dense(output_dim= 1))
LSTM_model_v2.add(Activation("linear"))
LSTM_model_v2.compile(loss="mse", optimizer= "rmsprop")
```

```python
history = LSTM_model_v2.fit(X_train, y_train, batch_size= 512, nb_epoch= 50, validation_split= 0.05)
```

### 5. Learning Curve

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='valid')
plt.title("Learning Curve")
plt.legend()
```

### 6. Predict

```python
predict_day = 10
```

```python
y_pred = []

for index in range(int(len(X_test) / predict_day)):
    current_frame = X_test[index * predict_day]
    predicted = []
    for day in range(predict_day):
        predicted.append(LSTM_model_v2.predict(current_frame[newaxis,:,:])[0,0])
        current_frame = current_frame[1:]
        current_frame = np.insert(current_frame, [period_day-1], predicted[-1], axis= 0)
    y_pred.append(predicted)
```

### 7. Visualize predict

```python
fig = plt.figure(figsize= (15, 5))
ax = fig.add_subplot(111)

ax.plot(y_test, label= "True Data")
for index, data in enumerate(y_pred):
    padding = [None for p in range(index * predict_day)]
    plt.plot(padding + data)
plt.title("LSTM Predict")
plt.legend()
```
